### Scatter plot that shows the relationship between the total fat and the cholesterol

```python
# Import pandas, a data processing and CSV file I/O library
import pandas as pd

# Import the Python graphing libraries Seaborn and Matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fast food nutrition dataset
ffn = pd.read_csv("fast_food_nutrition.csv") # the dataset is now a Pandas DataFrame

# Plotting the scatter plot using the columns total_fat and cholesterol
ffn.plot(kind="scatter", x="total_fat", y="cholesterol")
```

### Boxplot that shows the average number of calories that each meal has from different restaurants

```python
sns.set_theme(style="ticks", palette="pastel")
sns.set_context("paper")

# Plotting the boxplot using the columns restaurant and calories
sns.boxplot(x="restaurant", y="calories", data=ffn)
```

### Bar chart that shows the average number of calories by restaurant

```python
#Data preparation
restaurants_grouped = ffn[['restaurant', 'calories']].groupby('restaurant').mean().reset_index()

# Creating a bar chart that shows the average number of calories by restaurant
plt.bar(restaurants_grouped['restaurant'], restaurants_grouped['calories'])

# Setting the graph title
plt.title('Average number of calories by restaurant')

# Plotting the graph
plt.show()
```

### Pie chart that shows the number of items by restaurant

```python
# Data preparation
restaurants_grouped = ffn.groupby('restaurant')['item'].count().reset_index()

# Creating a pie chart that shows the number of items by restaurant
plt.pie(restaurants_grouped['item'], labels=restaurants_grouped['restaurant'])

# Setting the graph title
plt.title('Number of items by restaurant')

# Plotting the graph
plt.show()
```

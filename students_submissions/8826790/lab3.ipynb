```python
# Import all needed libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression

# Load the diabetes dataset as a numpy array using return_X_y=True
# Load x as a dataframe and y as a series using as_frame=True
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True, as_frame=True)

# Extract the BMI feature
diabetes_x_reshape = diabetes_X['bmi'].values.reshape(-1, 1)
diabetes_y_reshape = diabetes_y.values.reshape(-1, 1)

# Split the data into training and testing sets (20% Test data and 80% Training data)
x_train, x_test,y_train,y_test = train_test_split(diabetes_x_reshape,diabetes_y_reshape,test_size=0.20)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(x_train,y_train)

# Make predictions
predicted_values = model.predict(x_test)

# Calculate the score (R^2) of the linear regression model
score = model.score(x_test,y_test)
print(f'R-squared score: {score:.2f}')

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, predicted_values)
print(f'Mean squared error: {mse:.2f}')

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, predicted_values)
print(f'Mean absolute error: {mae:.2f}')

# Get the intercept and slope of the linear regression model
print(f'intercept: {model.intercept_[0]:.2f}')
print(f'slope: {model.coef_[0][0]:.2f}')

# -------------------------------------------------------------------
# Training Data Graph
# -------------------------------------------------------------------
# Plot training data
plt.scatter(x_train, y_train, color='black', label='Training Data')

# Plot the regression line
plt.plot(x_train, model.predict(x_train), color='blue', label='Regression Line')

# Add labels and title
plt.title('BMI vs. Diabetes Progression Linear Regression Model (Training Data)')
plt.xlabel('BMI')
plt.ylabel('Diabetes Progression')
plt.legend()

# Show plot
plt.show()

# -------------------------------------------------------------------
# Test Data Graph
# -------------------------------------------------------------------
# Plot test data
plt.scatter(x_test, y_test, color='purple', label='Test Data')

# Plot the regression line
plt.plot(x_test, model.predict(x_test), color='blue', label='Regression Line')

# Add labels and title
plt.title('BMI vs. Diabetes Progression Linear Regression Model (Test Data)')
plt.xlabel('BMI')
plt.ylabel('Diabetes Progression')
plt.legend()

# Show plot
plt.show()

# -------------------------------------------------------------------
# Training and Test Data Graph
# -------------------------------------------------------------------
# Plot training data
plt.scatter(x_train, y_train, color='black', label='Training Data')

# Plot test data
plt.scatter(x_test, y_test, color='purple', label='Test Data')

# Plot the regression line
plt.plot(x_train, model.predict(x_train), color='blue', label='Regression Line')

# Add labels and title
plt.title('BMI vs. Diabetes Progression Linear Regression Model (Training and Test Data)')
plt.xlabel('BMI')
plt.ylabel('Diabetes Progression')
plt.legend()

# Show plot
plt.show()
```

##### Based on the metrics calculated:
##### Mean Squared Error (MSE) measures the average square difference between the predicted values and the actual ones.
##### Mean Absolute Error (MAE) measures the average absolute difference between the predicted values and the actual ones.
##### These 2 measures are based on the units of the dependent variable (diabetes_y) and in this case I consider that more information is needed, because the units of this variable are not specified, so it cannot be concluded if these values are high or low. 

##### Coefficient of Determination represents the proportion of the variance in the dependent variable (diabetes progression) that is predictable from the independent variable (BMI). Considering that a score of 1.0 indicates a perfect fit, this model captures only a moderate amount of the variability in diabetes progression based on BMI.

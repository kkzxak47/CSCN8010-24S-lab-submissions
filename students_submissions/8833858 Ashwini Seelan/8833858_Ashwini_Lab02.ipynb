# Boston Housing Dataset Visualization

This notebook demonstrates different types of visualizations using the Boston Housing dataset. I have Used  Matplotlib and Seaborn for creating these visualizations.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
boston = pd.read_csv('housing.csv')
```

```python
print(boston.head())
print(boston.columns)
```

## 1. Histogram of House Prices

This histogram shows the distribution of median house values in the dataset.

```python
# Creating the graph1
plt.figure(figsize=(8, 6))
plt.hist(boston['MEDV'], bins=50)
plt.xlabel('House Price')
plt.ylabel('Frequency')
plt.title('Distribution of House Prices')
plt.show()
```

## 2. Correlation Matrix

This heatmap shows the correlation between different features in the dataset.


```python
# Creating the graph2
plt.figure(figsize=(8, 6))
sns.heatmap(boston.corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

## 3. Scatter Plot of House Prices vs. Number of Rooms

This scatter plot shows the relationship between the average number of rooms and house prices.


```python
# Creating the graph3
plt.figure(figsize=(8, 6))
sns.scatterplot(x='RM', y='MEDV', data=boston)
plt.xlabel('Number of Rooms')
plt.ylabel('House Price')
plt.title('Scatter Plot of House Prices vs. Number of Rooms')
plt.show()
```

## 4. Box Plot of House Prices by PTRATIO

This box plot shows the distribution of house prices based on the pupil-teacher ratio by town.


```python
# Creating the graph4
plt.figure(figsize=(8, 6))
sns.boxplot(x='PTRATIO', y='MEDV', data=boston)
plt.xlabel('Pupil-Teacher Ratio by Town')
plt.ylabel('House Price')
plt.title('Box Plot of House Prices by PTRATIO')
plt.show()
```

## 5. Pair Plot of Selected Features

This pair plot shows the pairwise relationships between selected features in the dataset.


```python
# Creating the graph5
sns.pairplot(boston[['RM', 'LSTAT', 'PTRATIO', 'MEDV']])
plt.show()
```

## 6. Line Plot of House Prices vs. Percentage of Lower Status Population

This line plot shows the relationship between the percentage of lower status of the population ('LSTAT') and the median house value ('MEDV').

```python
# Creating the graph6
plt.figure(figsize=(10, 6))
sns.lineplot(x='LSTAT', y='MEDV', data=boston)
plt.xlabel('Percentage of Lower Status Population')
plt.ylabel('Median House Price')
plt.title('House Prices vs. Percentage of Lower Status Population')
plt.show()
```

## 7. Swarm Plot of House Prices by Average Number of Rooms

This swarm plot shows the distribution of house prices based on the average number of rooms per dwelling ('RM').


```python
# Creating the graph7
plt.figure(figsize=(10, 6))
sns.swarmplot(x='RM', y='MEDV', data=boston)
plt.xlabel('Average Number of Rooms')
plt.ylabel('House Price')
plt.title('House Prices by Average Number of Rooms')
plt.show()
```

# Univariate Linear Regression on the Scikit-Learn Diabetes Dataset


This notebook demonstrates how to perform a univariate linear regression on the BMI feature of the Diabetes dataset to predict disease progression. 

```python
# Step 1: Frame the Problem
# The goal is to predict the disease progression one year after baseline using BMI (body mass index) as the predictor.


# Importing the necessary libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Step 2: Getting the Data
# Loading the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

```python
# Using only one feature (BMI - body mass index)
diabetes_X_bmi = diabetes_X[:, np.newaxis, 2]
```

```python
# Step 3: Exploring the Data
# Visualizing the relationship between BMI and disease progression
plt.scatter(diabetes_X_bmi, diabetes_y, color='blue')
plt.xlabel('BMI')
plt.ylabel('Disease progression')
plt.title('BMI vs Disease progression')
plt.show()
```

```python
# Step : Fine-tuning the Model
# Spliting the data into training and testing sets
diabetes_X_train, diabetes_X_test, diabetes_y_train, diabetes_y_test = train_test_split(diabetes_X_bmi, diabetes_y, test_size=0.2, random_state=42)
```

```python
# Creating a linear regression object
regr = LinearRegression()
```

```python
# Training the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)
```

```python
# Making predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)
```

```python
# The mean squared error
print('Mean squared error on training set:', mean_squared_error(diabetes_y_train, regr.predict(diabetes_X_train)))
print('Mean squared error on test set:', mean_squared_error(diabetes_y_test, diabetes_y_pred))
```

```python
# Step 7: Presenting the solution
# The coefficients
print('Coefficient (slope):', regr.coef_[0])
print('Intercept:', regr.intercept_)

# The mean squared error
print('Mean squared error on training set:', mean_squared_error(diabetes_y_train, regr.predict(diabetes_X_train)))
print('Mean squared error on test set:', mean_squared_error(diabetes_y_test, diabetes_y_pred))
```

The coefficient (slope) and intercept give us the parameters of the linear model.
The mean squared error provides a measure of the loss, indicating how well the model performs.


```python
# Ploting the result
plt.scatter(diabetes_X_train, diabetes_y_train, color='blue', label='Training data')
plt.scatter(diabetes_X_test, diabetes_y_test, color='green', label='Testing data')
plt.plot(diabetes_X_test, diabetes_y_pred, color='red', linewidth=2, label='Linear regression line')
plt.xlabel('BMI')
plt.ylabel('Disease progression')
plt.title('BMI vs Disease progression with Linear Regression Line')
plt.legend()
plt.show()
```

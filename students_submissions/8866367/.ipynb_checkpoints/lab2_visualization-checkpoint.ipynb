```python
# First, we'll import pandas, a data processing and CSV file I/O library
import pandas as pd

# We'll also import seaborn, a Python graphing library
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
```

```python
heart = pd.read_csv("../8866367/heart.csv") # dataset to pandas dataframe
heart.head() #last results in dataset
```

```python
# Let's see how many examples we have of Resting ECG
heart["RestingECG"].value_counts()
```

```python
# scatterplot for age vs resting blood pressure
heart.plot(kind="scatter", x="Age", y="RestingBP", title="Age vs. Resting Blood Pressure")
```

```python
# using seaborn for the same plot
sns.scatterplot(x="Age", y="RestingBP", data=heart, size=5)
```

```python
#visualize the relationship between age and resting blood pressure while distinguishing between patients with and without heart disease.
g = sns.FacetGrid(heart, hue="HeartDisease", height=6)
g.map(plt.scatter, "Age", "RestingBP").add_legend()
plt.title("Age vs. Resting Blood Pressure Colored by Heart Disease")
plt.xlabel("Age")
plt.ylabel("Resting Blood Pressure")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="HeartDisease", y="Cholesterol", data=heart)
plt.title("Cholesterol Levels by Heart Disease Status")
plt.xlabel("Heart Disease")
plt.ylabel("Cholesterol")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="HeartDisease", y="Cholesterol", data=heart)
sns.stripplot(x="HeartDisease", y="Cholesterol", data=heart, jitter=True, edgecolor="gray")
plt.title("Cholesterol Levels by Heart Disease Status")
plt.xlabel("Heart Disease")
plt.ylabel("Cholesterol")
plt.show()
```

```python
sns.violinplot(x="HeartDisease", y="Cholesterol", data=heart)
```

```python
sns.pairplot(heart, hue='HeartDisease')
```

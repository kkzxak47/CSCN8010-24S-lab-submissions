# Lab3 - Univariate Linear Regression on the Scikit-Learn Diabetes Dataset


#### Step 1: Frame the Problem and Look at the Big Picture
**Objective**: Predict disease progression (a quantitative measure one year after baseline) using Body Mass Index (BMI) as the independent variable.

**Problem Type**: Supervised learning, regression task.

**Performance Metric**: Mean Squared Error (MSE) will be used to evaluate the model performance.

#### Step 2: Obtain the Data
We'll utilize Scikit-Learn's datasets module to load the necessary data.

```python
from sklearn import datasets
import numpy as np

# Load the dataset
data_x, data_y = datasets.load_diabetes(return_X_y=True)

# use the Body mass index
index_col = 2
data_x_bmi = data_x[:, np.newaxis, index_col]
```

### Step 3: Analyze the Data for Insights
We'll create a plot to visualize the relationship between BMI and disease progression.

```python
import matplotlib.pyplot as plt

plt.scatter(data_x_bmi, data_y)
plt.xlabel('body mass index')
plt.ylabel('Progression of disease')
plt.title('body mass index vs Progression of disease')
plt.show()
```

#### Step 4: Prepare the Data
The data is already clean and ready for modeling, so no additional preparation is needed.

#### Step 5: Select Potential Models
Given our focus on univariate linear regression, we will use this model.

#### Step 6: Optimize the Model
Divide the data into training and testing sets, train the model, and assess its performance.


```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# spliting data into test and train data
data_x_train, data_x_test, data_y_train, data_y_test = train_test_split(data_x_bmi, data_y, test_size=0.2, random_state=42)

# creating linear regression model
model = LinearRegression()

# train the models
model.fit(data_x_train, data_y_train)

# predict
data_y_pred = model.predict(data_x_test)

# mean squared error
mean_sq_err = mean_squared_error(data_y_test, data_y_pred)

print(f'Model slope (coefficient): {model.coef_[0]}')
print(f'Model intercept: {model.intercept_}')
print(f'Mean Squared Error: {mean_sq_err}')
```

#### Step 7: Present the Results
Plot the regression line using both the training and test data to show how well the model fits.


```python
plt.figure(figsize=(12, 5))

# train data
plt.subplot(1, 2, 1)
plt.scatter(data_x_train, data_y_train, color='blue')
plt.plot(data_x_train, model.predict(data_x_train), color='red', linewidth=2)
plt.title('Train the data with Linear Regression Line')
plt.xlabel('Body Mass Index')
plt.ylabel('Progression of disease')

# test data
plt.subplot(1, 2, 2)
plt.scatter(data_x_test, data_y_test, color='blue')
plt.plot(data_x_test, data_y_pred, color='red', linewidth=2)
plt.title('Test the data with Linear Regression Line')
plt.xlabel('Body Mass Index')
plt.ylabel('Progression of disease')

plt.show()
```

**Model Parameters:** The slope and intercept of the linear regression model provide the relationship between BMI and disease progression.

**Model Performance:** The Mean Squared Error (MSE) on the test set gives an indication of the model's accuracy. A lower MSE indicates a better fit.

**Fit Visualization:** The regression line plotted against the training and test data shows how well the model captures the trend in the data.


Given the nature of the data and the simplicity of the model (univariate linear regression), it's important to assess the goodness of fit both visually and through performance metrics. If the relationship between BMI and disease progression is linear, this model should perform well. However, if the relationship is more complex, a more sophisticated model may be required.

#### Importing Required Packages

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

#### Load The Dataset

```python
netflix_data = pd.read_csv("C:\\Users\\pcpra\\Downloads\\netflix_titles.csv\\netflix_titles.csv" ,encoding='ISO-8859-1')
```

### 1.Understanding And Visualizing The Data Using Matplotlib

#### 1.1 Distribution of Content by Country (Bar Chart):

```python
plt.figure(figsize=(12, 6))
top_countries = netflix_data['country'].value_counts().head(10)
top_countries.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with Most Content on Netflix')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

#### 1.2 Trend of Content Releases Over the Years (Line Plot):


```python
plt.figure(figsize=(12, 6))
netflix_data['release_year'].value_counts().sort_index().plot(kind='line', marker='o', color='orange')
plt.title('Trend of Content Releases Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### 3.Understanding And Visualizing Using Seaborn

#### 3.1 Distribution of Duration for Movies (Histogram):

```python
# Filter the DataFrame to include only movies
netflix_movies = netflix_data[netflix_data['type'] == 'Movie']


plt.figure(figsize=(10, 6))
sns.histplot(netflix_movies['duration'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Movie Durations on Netflix')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.show()
```

#### 3.2 Trend of Content Releases Over the Months (Bar Plot):

```python
netflix_data['date_added'] = pd.to_datetime(netflix_data['date_added'])

# Extracting month from the 'date_added' column
netflix_data['month_added'] = netflix_data['date_added'].dt.month

# Trend of Content Releases Over the Months (Bar Plot)
plt.figure(figsize=(10, 6))
sns.countplot(x='month_added', data=netflix_data, palette='muted')
plt.title('Trend of Content Releases Over the Months')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

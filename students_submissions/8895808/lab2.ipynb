```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("D:\GIT_REPO\CSCN8010-24S-lab-submissions\students_submissions\8895808\WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
plt.figure(figure=(5, 7))
sns.lineplot(x='BusinessTravel', y='Attrition', data=data, marker='o')
plt.title('Attrition rate in Companies')
plt.xlabel('Business Travel')
plt.ylabel('Attrition')
plt.tight_layout()
plt.show()
```

This graphs shows the attrition rate (Yes/No) in companies basis on the travel frequency of the employees according to HR prediction.

```python
# Bar Plot: Total Distributed Vaccines by Location
plt.figure(figure=(12, 7))
sns.barplot(x='MaritalStatus', y='TotalWorkingYears', data=data)
plt.title('Total working Years according to Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Total Working Years')
plt.tight_layout()
plt.show()
```

This graphs shows the total working years given by the employees as per their marital status according to HR prediction.  

```python
plt.figure(figsize=(14, 7))
sns.scatterplot(x='Age', y='JobRole', data=data, hue='Department', size='Attrition', sizes=(20, 200))
plt.title('Attrition Rate of employees')
plt.xlabel('Age')
plt.ylabel('Job Role')
plt.legend(title='Department', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

This scatter plot shows the relationship between age and job role of the employees in a company and how attrition rate is predicted according to the job roles in a particular department.

```python

plt.figure(figure=(12, 7))
sns.histplot(data['MonthlyIncome'], bins=20, kde=True)
plt.title('Monthly Income')
plt.xlabel('MonthlyIncome')
plt.ylabel('JobRole')
plt.tight_layout()
plt.show()
```

This histogram illustrates the monthly income of employees given in a dataset.

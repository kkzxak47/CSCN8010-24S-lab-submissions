# Scatter Plot with Random Data (matplotlib)


```python
import matplotlib.pyplot as plt
import numpy as np

# Use the '_mpl-gallery' style for the plot
plt.style.use('_mpl-gallery')

# Set the seed for reproducibility of random data
np.random.seed(3)

# Generate random data for x and y coordinates
x = 4 + np.random.normal(0, 2, 24)  # 24 random values centered around 4
y = 4 + np.random.normal(0, 2, len(x))  # Same length as x

# Generate random sizes and colors for the scatter plot points
sizes = np.random.uniform(15, 80, len(x))  # Sizes between 15 and 80
colors = np.random.uniform(15, 80, len(x))  # Colors between 15 and 80

# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Create a scatter plot with x, y coordinates, sizes, and colors
ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

# Set the x and y axis limits and tick marks
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),  # x-axis from 0 to 8 with ticks from 1 to 7
       ylim=(0, 8), yticks=np.arange(1, 8))  # y-axis from 0 to 8 with ticks from 1 to 7

# Display the plot
plt.show()
```

# Histogram of Data Distribution (matplotlib)

```python
import matplotlib.pyplot as plt
import numpy as np

# Set the plot style
plt.style.use('_mpl-gallery')

# Generate data
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Create a histogram with specified bins, line width, and edge color
ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

# Set the x and y axis limits and ticks
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

# Display the plot
plt.show()
```

# Stacked Histogram of Diamond Prices by Cut (Seaborn)

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

# Set the seaborn theme
sns.set_theme(style="ticks")

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create a figure and axis
f, ax = plt.subplots(figsize=(7, 5))

# Remove the top and right spines from the plot
sns.despine(f)

# Create a stacked histogram of diamond prices based on cut
sns.histplot(
    diamonds,
    x="price", hue="cut",  # Plot price against cut with different colors for each cut
    multiple="stack",  # Stack the histograms
    palette="light:m_r",  # Use a light to medium red color palette
    edgecolor=".3",  # Set edge color to gray
    linewidth=.5,  # Set line width for edges
    log_scale=True,  # Use a logarithmic scale for the x-axis
)

# Format x-axis ticks using ScalarFormatter
ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())

# Set custom x-axis ticks
ax.set_xticks([500, 1000, 2000, 5000, 10000])

# Show the plot
plt.show()
```

# Interactive Line Plot: Life Expectancy in Canada (Plotly Express)


```python
import plotly.express as px  # Import Plotly Express for interactive plotting

# Load the Gapminder dataset and filter for Canada
df = px.data.gapminder().query("country=='Canada'")

# Create a line plot using Plotly Express
fig = px.line(df, x="year", y="lifeExp", title='Life expectancy in Canada')

# Show the interactive plot
fig.show()
```

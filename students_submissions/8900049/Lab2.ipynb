```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("speed_data_data.csv")
```

# **Bar plot**
Below  bar graph shows the distribution of individuals across different careers.The x-axis represents different career options, and the y-axis represents the number of entries  in each career.

```python
career_counts = df['career'].value_counts()
plt.figure(figsize=(50, 6))
career_counts.plot(kind='bar', color='green')
plt.xlabel('Career')
plt.ylabel('Number of Entries')
plt.title('Distribution of Individuals Across Different Careers')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# **Scatter plot**
Below scatter graph shows the relationship between age and income among individuals in the dataset.Each point represents an individual, with their age on the x-axis and income on the y-axis.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['income'], color='blue', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Relationship between Age and Income')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

# **Histogram plot**
Below histogram graph shows the distribution of ages among individuals in the dataset.The x-axis represents age, and the y-axis represents the frequency (number of individuals) in each age bin.The bars represent the distribution of ages.

```python
# Histogram Plot
##The x-axis represents age, and the y-axis represents the frequency (number of individuals) in each age bin.
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=10, color='red', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title(' Frequency Distribution of Age')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# **Pie Chart**
Below pie chart shows the percentage of individuals having each attribute such as *attractiveness, sincerity, intelligence, fun, ambitiousness* and *shared interests*

```python
attributes = ["attr", "sinc", "intel", "fun", "amb", "shar", "like", "prob", "met"]
percentages = [(df[attr].count() / len(df)) * 100 for attr in attributes]
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=attributes, autopct='%1.1f%%')
plt.title('Percentage of Individuals Having Attributes')
plt.show()
```

# Lab 2 For Graph Visualization on Manga Dataset
### Dataset: https://www.kaggle.com/datasets/drahulsingh/best-selling-manga

```python
import pandas as pd
df = pd.read_csv('best-selling-manga.csv')
df.head()
```

# Lab 2 For Graph Visualization on Manga Dataset
### Dataset: https://www.kaggle.com/datasets/drahulsingh/best-selling-manga

#Now we import the library required for the First Graph. This is a bar chart showing the top 10 manga's by approximate sales.

```python
import matplotlib.pyplot as plt

# Select top 10 manga series by approximate sales
top10_sales = df.nlargest(10, 'Approximate sales in million(s)')

# Plot
plt.figure(figsize=(12, 8))
plt.barh(top10_sales['Manga series'], top10_sales['Approximate sales in million(s)'], color='skyblue')
plt.xlabel('Approximate Sales in Million(s)')
plt.title('Top 10 Manga Series by Sales')
plt.gca().invert_yaxis()
plt.show()
```

# Making Scatter plot using Seaborn to show number of volumes

```python
import seaborn as sns

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='No. of collected volumes', y='Approximate sales in million(s)', hue='Demographic', data=df, palette='viridis')
plt.title('Sales vs. Number of Collected Volumes')
plt.xlabel('Number of Collected Volumes')
plt.ylabel('Approximate Sales in Million(s)')
plt.legend(title='Demographic')
plt.show()
```

# Using MatPlotLIB for Pie Chart of Demographics 

```python

demographic_counts = df['Demographic'].value_counts()

colors = ['green', 'red']

plt.figure(figsize=(8, 8))
patches, texts, autotexts = plt.pie(demographic_counts, labels=demographic_counts.index, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Demographic Distribution of Manga Series')

# Add a legend
plt.legend(patches, demographic_counts.index, loc="best")


plt.show()
```

# Using Seaborn Heatmap for Average Sales per Publisher

```python
heatmap_data = df.pivot_table(values='Average sales per volume in million(s)', index='Publisher', columns='Demographic', aggfunc='mean')

# Plot
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm')
plt.title('Average Sales per Volume by Publisher and Demographic')
plt.show()
```

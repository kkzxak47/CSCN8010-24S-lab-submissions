```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
```

## Example 1

In this example, I'm utilizing Matplotlib and NumPy to showcase a sine wave...


```python
xpts= np.arange(2,10)
ypts=np.sin(xpts)

plt.plot(xpts,ypts,linestyle='dashed')

plt.title('Fig. 1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid()
plt.show()
```

<sub><sup>Above example, we named the diagram as Fig. 1 and plotted the sine wave along the y-axis.</sup></sub>

## Example 3

This code generates a scatter plot showing the relationship between age and cholesterol levels, with the color indicating the target variable, using data from a CSV file.

```python

df= pd.read_csv("heart.csv")
plt.scatter(x='age',y='chol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('Age vs. Cholesterol with Target')
plt.colorbar(label='Target')
plt.show()
```

## Example 4
We utilized Seaborn, a Python visualization library, to create the pairplot. Seaborn provides a high-level interface for drawing attractive and informative statistical graphics.


```python

# Load the data
df = pd.read_csv("heart.csv")

sns.pairplot(df.iloc[:, :3])  # Selecting only the first three columns
plt.show()
```

## Example 5

To visualize a scatter plot using seaborn with the first five columns of a DataFrame.

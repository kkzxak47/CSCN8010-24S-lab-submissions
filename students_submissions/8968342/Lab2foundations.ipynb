```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px
```

```python
file_path = 'D:\downolad 1\heart.csv'

# Read the CSV file
data = pd.read_csv("D:\downolad 1\heart.csv")

# Display the first few rows of the dataframe
print(data.head())

# Print the column names
print("Column names:", data.columns)
```

```python
# Age Distribution
if 'Age' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data['Age'], bins=10, kde=True, color='skyblue')
    plt.title('Age Distribution of Patients')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Gender Distribution
if 'Sex' in data.columns:
    fig = px.pie(data, names='Sex', title='Gender Distribution of Patients')
    fig.show()
```

```python
 # Cholesterol vs. Age
if 'Age' in data.columns and 'Cholesterol' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x='Age', y='Cholesterol', hue='Age', palette='viridis', s=100)
    plt.title('Cholesterol vs. Age')
    plt.xlabel('Age')
    plt.ylabel('Cholesterol')
    plt.show()
```

```python
# Boxplot of Ejection Fraction by Death Event
if 'Age' in data.columns and 'MaxHR' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=data, x='Age', y='MaxHR', palette='Set2')
    plt.title('Boxplot of MaxHRn by Age')
    plt.xlabel('Age')
    plt.ylabel('MaxHR')
    plt.show()
```

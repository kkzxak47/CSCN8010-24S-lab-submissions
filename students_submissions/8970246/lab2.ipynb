```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

Basic Graph

```python
xpoints = np.array([0, 1, 2, 6])
ypoints = np.array([0, 5, 4, 8])

#Creating a basic graph
plt.plot(xpoints, ypoints, marker = '.', markersize = 10)

#Title of the Graph
plt.title("A Basic Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
# show plot
plt.show()
```

Bar Chart using Matplotlib

```python
   
# creating the dataset
data = {'Python':99, 'ML_Lab':92, 'AI_Lab':80, 
        'Agile_Software':70}
programs = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(programs, values, color ='purple', 
        width = 0.4)
 
plt.xlabel("Courses offered")
plt.ylabel("Grades in Percentage")
plt.title("Student Grade")
plt.show()
```

Pie chart using Matplotlib

```python
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

#sales data in percentage
sales_data = [90, 95, 50, 98, 60, 40]
 
# Creating plot
fig = plt.figure(figsize=(10, 7))
plt.pie(sales_data, labels=cars, autopct='%1.1f%%')
plt.title("Pie chart representation of sales percentage for the list of cars")
 
# show plot
plt.show()
```

Pie chart Plot using Dataset csv file

```python
#datset from kaggle site
df = pd.read_csv('Student_Mental_Health.csv')
depression_counts = df['Do you have Depression?'].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(depression_counts, labels=depression_counts.index, autopct='%1.1f%%', colors=['skyblue', 'salmon'])
plt.title('Proportion of Depression Responses')
plt.show()
```

Density plot with Seaborn

```python
np.random.seed(0)
x = np.random.normal(loc=0, scale=1, size=1000)
y = np.random.normal(loc=0, scale=1, size=1000)
categories = ['A', 'B', 'C', 'D']
values = [25, 50, 75, 100]

# Density plot with Seaborn
plt.figure(figsize=(8, 5))
#
sns.kdeplot(x=x, y=y, cmap='plasma', fill=True)
plt.title('Density Plot with Seaborn')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
# show plot
plt.show()
```

Seaborn facet grids

```python
tips = sns.load_dataset('tips')

# created a FacetGrid for day vs total_bill

t = sns.FacetGrid(tips, col="day")

# plot histogram for total_bill in each day

t.map(sns.histplot, "total_bill")
```

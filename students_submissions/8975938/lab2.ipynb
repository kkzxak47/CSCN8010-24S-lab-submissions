## Reading dataset

```python
# First, we'll import pandas, a data processing and CSV file I/O library
import pandas as pd

# We'll also import seaborn, a Python graphing library
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)

# Next, we'll load the Iris flower dataset, which is in the "../input/" directory
hbr = pd.read_csv("HBR.csv") # the iris dataset is now a Pandas DataFrame

# Let's see what's in the iris data - Jupyter notebooks print the result of the last thing you do
hbr.head()
```

### Bar chart
Letâ€™s examine the age distribution, in other words, how many people are in each age group.

```python

# Let's see the age distribution
hbr["age"].value_counts()

# Plotting the distribution of ages
sns.histplot(hbr["age"], bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Box plot
Group by age and calculate the percentage distribution of cholesterol

```python
# Group by age and calculate the percentage distribution of cholesterol
bins = [29, 39, 49, 59, 69, 79]  # Define age bins
labels = ['30-39', '40-49', '50-59', '60-69', '70-79']  # Labels for the age bins
hbr['age_group'] = pd.cut(hbr['age'], bins=bins, labels=labels)

plt.figure(figsize=(10, 6))
sns.boxplot(x='age_group', y='chol', data=hbr)
plt.title('Cholesterol Levels by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Cholesterol')
plt.show()
```

### Count plot
Count Plot of Target by Sex

```python
sns.countplot(x='sex', hue='target', data=hbr)
plt.title('Count of Target by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Target')
plt.show()
```

### Scatter plot
Scatter Plot of Cholesterol vs Thalach (Maximum Heart Rate)

```python
sns.scatterplot(x='chol', y='thalach', hue='target', data=hbr, palette='viridis', s=100)
plt.title('Cholesterol vs Thalach (Maximum Heart Rate)')
plt.xlabel('Cholesterol')
plt.ylabel('Maximum Heart Rate (Thalach)')
plt.legend(title='Target')
plt.show()
```

### Bar plot
Bar Plot of Average Resting Blood Pressure by Chest Pain Type

```python
sns.barplot(x='cp', y='trestbps', data=hbr, ci=None, palette='muted')
plt.title('Average Resting Blood Pressure by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Average Resting Blood Pressure')
plt.show()
```

## Data Visualization

### 1. HISTOGRAM
The following visualization shows HISTOGRAM describing the Math, Reading, and Writing Score distributions

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 10)) 
fig.suptitle('Analyzing the Math, Reading, and Writing Score Distribution.', fontsize=20)

#Histogram for Math Score Distribution 
sns.histplot(scores, x="math score", ax=axes[0])
axes[0].title.set_text('Histogram for Math Score')

#Histogram for Reading Score Distribution with different colors for ranges of marks scored
sns.histplot(scores, x="reading score", ax=axes[1], hue="reading score", palette="Set1", legend=False)
axes[1].title.set_text('Histogram for Reading Score')

#Histogram for Writing Score Distribution
sns.histplot(scores, x="writing score", ax=axes[2])
axes[2].title.set_text('Histogram for Writing Score')

plt.show()
```

### 2. BOXPLOT (COMPARATIVE)

The visualization uses boxplot to compare the scores with gender
The boxplot explains the median score and the people who are outliers (whose score is below median)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 10)) 
fig.suptitle('Comparative boxplot for Scores Vs Gender', fontsize=20)
color_1 = ["#FF5733", "#33FF57"]
color_2 = ["#3357FF", "#ff9999"]
color_3 = ["#66b3ff", "#99ff99"]

#Boxplot for Math Score Distribution based on Gender
sns.boxplot(x="gender", y="math score", data = scores, ax=axes[0], hue = "gender", palette=color_1)
axes[0].title.set_text('Math Score Vs Gender')


#Boxplot for Reading Score Distribution based on Gender
sns.boxplot(x="gender", y="reading score", data = scores, ax=axes[1], hue = "gender", palette=color_2)
axes[1].title.set_text('Reading Score Vs Gender')


#Boxplot for Writing Score Distribution based on Gender
sns.boxplot(x="gender", y="writing score", data = scores, ax=axes[2], hue = "gender", palette=color_3)
axes[2].title.set_text('Writing Score Vs Gender')

plt.tight_layout()  # Adjust layout to fit the title properly
plt.show()
```

### 3.GROUPED BARCHART
The grouped barchart explains the parental level of education against gender.

```python
plt.figure(figsize= (12, 5))
#counts the observations in each categorical feature from the dataset
sns.countplot(x = 'parental level of education', data = scores, hue = 'gender')
plt.title('Gender Vs Parental Education')
plt.show()
```

### 4. PIE CHART
The pie chart explains the percentage of different ethnicity groups in the dataset

```python
#counting the number of ethnicity in dataset
features = scores['ethnicity'].value_counts().index
data = scores['ethnicity'].value_counts().values
color_chart = ["blue", "yellow", "red", "green", "violet"]
plt.figure(figsize=(7,7))
plt.pie(data, labels=features, colors=color_chart, autopct="%1.1f%%")
plt.title("Ethnicity Details")
plt.show()
```

### 5. SCATTERPLOT
The scatterplot explains the relationship between the scores, for instance student performing well in one subject may also excel in other subjects as well. To find that replationship scatterplots are used

```python
#Relationship between Reading and Writing Score
sns.scatterplot(x='reading score',y='writing score',data=scores)
```

```python
#Relationship between Math and Writing Score
sns.scatterplot(x='math score',y='writing score',data=scores)
```

```python
#Relationship between Math and Reading Score
scatter_plot = graph.Scatter(x=scores['math score'], y=scores['reading score'])
layout = graph.Layout(title='ScatterPlot', xaxis=dict(title='math score'), yaxis=dict(title='reading score'))
fig = graph.Figure(data=[scatter_plot], layout=layout)
io.show(fig)
```

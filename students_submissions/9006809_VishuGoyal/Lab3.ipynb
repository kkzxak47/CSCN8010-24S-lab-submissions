```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

##### Loading dataset

```python
data_X, data_y = datasets.load_diabetes(return_X_y=True)
```

#### Fetching bmi (body mass index) feature from dataset

```python
data_bmi = data_X[:, np.newaxis, 2]

#Splitting data into two parts Test data and train data
data_bmi_train, data_bmi_test, data_y_train, data_y_test = train_test_split(data_bmi, data_y, test_size=0.3)
```

#### Training Regression Model using above dataset

```python

reg_model = LinearRegression()
reg_model.fit(data_bmi_train, data_y_train)

pred_data_y_test = reg_model.predict(data_bmi_test)

# Plotting Testing data
plt.scatter(data_bmi_test, data_y_test, color='blue', label="Testing Data")
plt.plot(data_bmi_test, pred_data_y_test, color='red', label="Regression Line")
plt.xlabel('BMI (Body Mass Index)')
plt.ylabel('Disease Progression')
plt.title('Univariate Linear Regression')
plt.legend()

```

#### Slope and Intercept 

```python
slope = reg_model.coef_
intercept = reg_model.intercept_

print("Slope:", slope)
print("Intercept:", intercept)
```

#### Mean Squared Error

```python
pred_data_y_train = reg_model.predict(data_bmi_train)

mse_train = mean_squared_error(data_y_train, pred_data_y_train)
mse_test = mean_squared_error(data_y_test, pred_data_y_test)

print(f"Mean squared error")
print(f"Training : {mse_train:.2f}")
print(f"Testing : {mse_test:.2f}\n")

print(f"The difference between the Training and Test error : {(mse_train - mse_test):.2f}")
```

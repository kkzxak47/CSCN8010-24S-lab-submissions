```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns 

```

1. Bar chart of Iris Dataset - Iris Species Vs Sepal Length

```python
#Importing Iris dataset
data = pd.read_csv("Iris.csv")
#x axis data
xaxis = data.Species
#y axis data
yaxis = data["SepalLengthCm"]
#labelling x axis
plt.xlabel("\n x axis = Species of Iris",color = "red")
#labelling y axis
plt.ylabel("\n y axis = Sepal Length (cm)",color = "red")
#labelling title
plt.title("Iris Species Vs Sepal Length" , loc = "left", color = "Blue")
#creates the plot
plt.bar(xaxis, yaxis , color = "green", width = 0.5)
#show the plot
plt.show()

```

2. Simple Line Chart to show mathematical expression  X Vs Y

```python
fig, pi = plt.subplots()
# x axis
x = np.linspace(0, 20)
# y axis
y = (x + x)
#labelling x
plt.xlabel("\n x axis")
#labelling y
plt.ylabel("\n y axis")
#labelling title
plt.title("x Vs y" , loc = "left", color = "Blue")
# creates the plot
pi.plot(x, y, color = "green") 
#show plot               
plt.show() 
```

3. Scatter Plot for Continents Ranks Vs Population (2022)

```python

#importing dataset
data1 = pd.read_csv("world_population.csv")

# Group by continent and sum the populations
population = data1.groupby('Continent')['2022 Population'].sum().reset_index()

#creating plot
sns.relplot(
    data=data1,
    x=population["2022 Population"],y='Rank',hue="Continent"
    
)

#labelling title
plt.title("Continents Ranks Vs Population (2022)" , loc = "left", color = "Blue")
```

4. Pie Chart for Population Distribution by Continents

```python

#importing dataset
df = pd.DataFrame(data1)

# Group by continent and sum the populations
population = df.groupby('Continent')['2022 Population'].sum().reset_index()

# Plot a pie chart
plt.figure(figsize=(10, 7))
plt.pie(population['2022 Population'], labels=population['Continent'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set3'))

# Add a title
plt.title('Population Distribution by Continents',color = "Red")

# Display the plot
plt.show()
```

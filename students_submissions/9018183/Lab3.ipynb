### Univariate Linear Regression on the Scikit-Learn Diabetes Dataset ###

```python
# Necessary libraries
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
# Load the diabetes dataset
d_X, d_y = datasets.load_diabetes(return_X_y=True)
print("Diabetes x :" ,d_X)
print("Diabetes y :" ,d_y)
# Select only the BMI feature (index 2)
d_X_bmi = d_X[:, np.newaxis, 2]
```

```python
# Splitting data to training and test sets
d_X_train, d_X_test, d_y_train, d_y_test = train_test_split(
    d_X_bmi, d_y, test_size=0.2, random_state=42)

# Creating linear regression model
model = LinearRegression()

# Training the model
model.fit(d_X_train, d_y_train)
```

```python
# Predictions
d_y_train_pred = model.predict(d_X_train)
d_y_test_pred = model.predict(d_X_test)
print("Diabetes y train data",d_y_train_pred)
print("Diabetes y test data",d_y_test_pred)
```

```python
# Mean squared error
train_mse = metrics.mean_squared_error(d_y_train, d_y_train_pred)
test_mse = metrics.mean_squared_error(d_y_test, d_y_test_pred)

print(f"\nTrain data mean squared error: ",train_mse)
print(f"\nTest data mean squared error: ",test_mse)
```

```python
# Graph of training data, test data, and the model
plt.scatter(d_X_train, d_y_train, color='Red', label='Training data')
plt.scatter(d_X_test, d_y_test, color='green', label='Test data')
plt.plot(d_X_bmi, model.predict(d_X_bmi), color='black', linewidth=2, label='Regression line')
plt.xlabel('Body Mass Index (BMI)')
plt.ylabel('Disease Progression')
plt.title('BMI vs Disease Progression')
plt.legend()
plt.show()

# Model parameters
print(f"Slope : {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")
```

 The graph is clearly capturing the overall trend of increasing disease progression with higher BMI values.
